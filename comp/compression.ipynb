```python
# imports
%matplotlib inline
import pandas
import matplotlib.pyplot as plt
import numpy as np

# figure size
plt.rcParams['figure.figsize'] = (15.0, 8.0)
font = {'family' : 'latex',
        'weight' : 'bold',
        'size'   : 26}
plt.rcParams['text.usetex'] = True

plt.rc('font', **font)
```

```python
prioNode0 = pandas.read_csv("log/priorityNode0.csv", index_col=False, header=0)
#prio.head()

fig, ax1 = plt.subplots()
plt.grid()
#plt.xlim([0,5])

ax1.set_xlabel(r"Sensitivity")
ax1.set_ylabel(r"Average Error")
ax1.set_ylim([0.0,0.05])
a = ax1.plot(prioNode0.sensitivity, prioNode0.priorityError, label='Average Error', color='red', ls="-", lw=3)

ax2 = ax1.twinx()
ax2.set_ylabel('Compression Ratio')
b = ax2.plot(prioNode0.sensitivity, prioNode0.compressionRatio, label='Compression Ratio', color='blue', ls="-", lw=3)


plt.rc('text', usetex=True)
plt.rc('font', family='serif')

plts = a + b
labs = [l.get_label() for l in plts]
ax1.legend(plts, labs, loc='upper center', bbox_to_anchor=(0.5, -0.10), frameon=False,
          fancybox=False, shadow=False, ncol=3, markerscale=1, numpoints=1)

plt.savefig('res/errorVsRatioPrioNode0.eps', bbox_inches='tight')
plt.savefig('res/errorVsRatioPrioNode0.pdf', bbox_inches='tight')
plt.savefig('res/errorVsRatioPrioNode0.png', bbox_inches='tight')
```
